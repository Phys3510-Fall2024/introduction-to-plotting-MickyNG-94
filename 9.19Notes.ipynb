```python
# import needed packages/modules/libraries

import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
```

```python
# define the potential energy function

def potential_energy(Q1, Q2, K1=1.0, K2=1.0, C=1):
    return 0.5*K1*Q1**2 + 0.5*K2*Q2**2 + C* Q2*Q1**2 #follow the readme.md file

#test potential energy

potential_energy(1, 0)
```

```python
#define meshes in q1 and Q2
# testing numbers
#came out wonky; he has a 3x3 matrix result

q1 = np.linspace(-1, 1, 3)
q2 = q1
Q1, Q2 = np.meshgrid(Q1, Q2)

potential_energy(Q1, Q2)
```
